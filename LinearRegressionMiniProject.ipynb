```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#import necc. libraries and modules
#numpy for numerical computations and array manipulations
#matplotlib for data visualisation
#train_test_split for data splitting
#linearregreesion for modelling
#mean_sqaured_error forassesing model accuracy from scikit learn
```

```python
np.random.seed(0)
X = 2*np.random.rand(100,1)
x = 3*np.random.rand(100,1)
Y = 4+3*X + np.random.randn(100,1)

print(X[0:9])
print(x[0:9])

#np.random.rand func. generates a 2d number array with 100rows,1columns containing random no. bwt. 0 and 1
#np.random.randn func. generates a array of numbers from a standard distribution, avr.=0, std.deviat.=1
```

```python
X_train, X_test, Y_train, Y_test, x_train, x_test = train_test_split(X,Y,x,test_size = 0.2, random_state = 42)
#test_size=0.2 means u r splitting the data such that 20% is for testing, 80% is for training
```

```python
plt.scatter(X_train, Y_train, color = 'red')
plt.scatter(x_train, Y_train, color = 'green')
plt.title("BigBoobs")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
#standard stuff to make the graph
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, Y_train)
lin_reg.fit(x_train, Y_train)
#start linear regression model w machine leanring r.s bwt. X and Y, hence X_train and Y_train
```

```python
Y_pred = lin_reg.predict(X_test)
y_pred = lin_reg.predict(x_test)
#machine is used to make predictions ontesting data that is X_test
```

```python
mse = mean_squared_error(Y_test, Y_pred)
mse2 = mean_squared_error(Y_test, y_pred)
print("Mean Squared Error RAHH", mse)
print("ONGXINRAN IS PRETTY", mse2)
#code calcaulates the mean_squared_error bwt. the test data and the predicted data
#to provide a quantitative measure of how well model fits data
```

```python
plt.scatter(X_train, Y_train, color = 'red')
plt.scatter(x_train, Y_train, color = 'green')
plt.plot(X_test, Y_pred, color = 'blue')
plt.plot(x_test, y_pred+1, color = 'yellow')
plt.title("BigBoobs")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
#standard stuff to make the graph
```
